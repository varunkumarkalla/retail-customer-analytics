```python
import pandas as pd
import numpy as np
```

```python
file_path = "../Dataset/online_retail_II.xlsx"
```

```python
df_2009 = pd.read_excel(
    file_path,
    sheet_name="Year 2009-2010"
)

df_2010 = pd.read_excel(
    file_path,
    sheet_name="Year 2010-2011"
)
```

```python
df = pd.concat(
    [df_2009, df_2010],
    ignore_index=True
)
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sort_values(
    ascending=False
)
```

```python
df[
    df["Customer ID"].isnull()
].head()
```

```python
df = df.dropna(
    subset=["Customer ID"]
)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df["Quantity"].describe()
```

```python
df[
    df["Quantity"] < 0
]
```

```python
len(
    df[
        df["Quantity"]<0
    ]
)
```

```python
sales_df = df[
    df["Quantity"] > 0
]
```

```python
sales_df.shape
```

```python
sales_df["Price"].describe()
```

```python
sales_df[
    sales_df["Price"] <= 0
]
```

```python
sales_df = sales_df[
    sales_df["Price"] > 0
]
```

```python
sales_df["Revenue"] = (
    sales_df["Quantity"]
    *
    sales_df["Price"]
)
```

```python
sales_df[
    ["Quantity",
     "Price",
     "Revenue"]
].head()
```

```python
sales_df["InvoiceDate"] = pd.to_datetime(
    sales_df["InvoiceDate"]
)
```

```python
sales_df["Year"] = (
    sales_df["InvoiceDate"]
    .dt.year
)
```

```python
sales_df["Month"] = (
    sales_df["InvoiceDate"]
    .dt.month
)
```

```python
sales_df["Day"] = (
    sales_df["InvoiceDate"]
    .dt.day
)
```

```python
sales_df.info()
```

```python
sales_df.to_csv(
    "../Dataset/clean_retail.csv",
    index=False
)
```
