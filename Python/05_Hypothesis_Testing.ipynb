```python
import pandas as pd
import numpy as np

from scipy.stats import (
    ttest_ind,
    f_oneway,
    chi2_contingency,
    shapiro,
    levene,
    mannwhitneyu
)

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales_df = pd.read_csv(
    "../Dataset/clean_retail.csv"
)
```

```python
sales_df.head()
```

```python
sample_revenue = (
    sales_df["Revenue"]
    .sample(
        5000,
        random_state=42
    )
)
```

```python
stat, p = shapiro(
    sample_revenue
)

print("Statistic:", stat)
print("P-value:", p)
```

```python
uk = sales_df[
    sales_df["Country"]=="United Kingdom"
]["Revenue"]

germany = sales_df[
    sales_df["Country"]=="Germany"
]["Revenue"]

stat,p = levene(
    uk,
    germany
)

print("Statistic:", stat)
print("P-value:", p)
```

```python
uk = sales_df[
    sales_df["Country"]
    == "United Kingdom"
]["Revenue"]

germany = sales_df[
    sales_df["Country"]
    == "Germany"
]["Revenue"]

stat, p = ttest_ind(
    uk,
    germany,
    equal_var=False
)

print("Statistic:", stat)
print("P-value:", p)
```

```python
print("UK Mean:",
      uk.mean())

print("Germany Mean:",
      germany.mean())
```

```python
uk = sales_df[
    sales_df["Country"]=="United Kingdom"
]["Revenue"]

germany = sales_df[
    sales_df["Country"]=="Germany"
]["Revenue"]

france = sales_df[
    sales_df["Country"]=="France"
]["Revenue"]
```

```python
from scipy.stats import f_oneway

stat, p = f_oneway(
    uk,
    germany,
    france
)

print("Statistic:", stat)
print("P-value:", p)
```

```python
table = pd.crosstab(
    sales_df["Country"],
    sales_df["Month"]
)

table.head()
```

```python
from scipy.stats import chi2_contingency

stat, p, dof, expected = chi2_contingency(
    table
)

print("Statistic:", stat)
print("P-value:", p)
print("Degrees of Freedom:", dof)
```

```python
stat, p = mannwhitneyu(
    uk,
    germany
)

print("Statistic:", stat)
print("P-value:", p)
```
