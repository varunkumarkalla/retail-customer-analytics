```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sales_df = pd.read_csv(
    "../Dataset/clean_retail.csv"
)
```

```python
sales_df.head()
```

```python
sales_df.shape
```

```python
sales_df["Revenue"].sum()
```

```python
total_revenue = sales_df["Revenue"].sum()

total_revenue
```

```python
sales_df["Customer ID"].nunique()
```

```python
sales_df["StockCode"].nunique()
```

```python
top_products = (
    sales_df
    .groupby("Description")
    ["Revenue"]
    .sum()
    .sort_values(
        ascending=False
    )
)
```

```python
top_products.head(10)
```

```python
top_products.head(10).plot(
    kind="bar"
)

plt.title(
    "Top 10 Products by Revenue"
)

plt.show()
```

```python
plt.figure(
    figsize=(10,6)
)

top_products.head(10).plot(
    kind="bar"
)

plt.title(
    "Top 10 Products by Revenue"
)

plt.tight_layout()

plt.savefig(
    "../Images/top_products.png"
)

plt.show()
```

```python
top_customers = (
    sales_df
    .groupby("Customer ID")
    ["Revenue"]
    .sum()
    .sort_values(
        ascending=False
    )
)
```

```python
top_customers.head(10)
```

```python
top_customers.head(10).plot(
    kind="bar"
)

plt.title(
    "Top Customers"
)

plt.show()
```

```python
plt.figure(
    figsize=(10,6)
)

top_customers.head(10).plot(
    kind="bar"
)

plt.title(
    "Top 10 Customers by Revenue"
)

plt.tight_layout()

plt.savefig(
    "../Images/top_customers.png"
)

plt.show()
```
