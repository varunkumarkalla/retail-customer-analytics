```python
plt.figure(figsize=(10,6))

sns.histplot(
    sales_df["Revenue"],
    bins=50
)

plt.title(
    "Revenue Distribution"
)

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    sales_df["Revenue"],
    bins=50
)

plt.title(
    "Revenue Distribution"
)

plt.tight_layout()

plt.savefig(
    "../Images/revenue_distribution.png"
)

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    x=sales_df["Revenue"]
)

plt.title(
    "Revenue Outlier Analysis"
)

plt.show()
```
